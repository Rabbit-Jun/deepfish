```python
import numpy as np
import pickle
import tensorflow as tf

def get_data():
    mnist = tf.keras.datasets.mnist
    (x_train, t_train), (x_test, t_test) = mnist.load_data()
    return x_test, t_test

x, t = get_data()
x = x.reshape(-1,784) # ndim =1 이 784가 되고 ndim = 0는 자동으로 하게 만들어줌
print(f'x shape: {x.shape}')
print(f't shape: {t.shape}')

print(f'len x: {len(x)}')
print(f'len x[0]: {len(x[0])}')
print(f'len x[1]: {len(x[1])}')
def sigmoid(x):
    return 1/(1+np.exp(-x))

def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))  # axis=1로 각 행의 최대값을 사용
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)  # sum도 axis=1로 설정



def init_network():
    with open("./sample_weight.pkl", 'rb') as f:
        network = pickle.load(f)
    return network

network = init_network()

def predict(network, x):
    w1, w2, w3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, w1) + b1
    print(f'a1: {a1.shape}')
    z1 = sigmoid(a1)

    print(f'z1: {z1.shape}')
    a2 = np.dot(z1, w2) + b2
    
 
    print(f'a2: {a2.shape}')

    z2 = sigmoid(a2)

    
    print(f'z2: {z2.shape}')

    a3 = np.dot(z2, w3) + b3
    print(a3.shape)
    y = softmax(a3)
    print(f'y: {y}')
    return y


print(type(predict(network, x)))
```

# 배치

- 데이터셋을 묶음으로 처리하는 것
- 배치를 하면 연산속도가 빨라진다 (하드웨어(gpu)의 병렬처리 능력 사용, 메모리에 i/o의 대기시간을 줄이고 데이터를 미리 로드할 수있고 연산 중에 cpu나 gpu를 연속적으로 수행할 수 있기 때문)

```python
def predict(network, x):
    w1, w2, w3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, w1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, w2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, w3) + b3
    y = softmax(a3)

    return y

x, t = get_data()
x = x.reshape(-1,784)
print(x[0].shape)
network = init_network()
print(len(x))
batch_size = 100 # 배치 크기
accuracy_cnt = 0

for i in range(0, len(x), batch_size): # 100개씩 묶어서 100번 돌림
    x_batch = x[i:i+batch_size]
    print(f'x_batch: {x_batch.shape}')
    y_batch = predict(network, x_batch)
    # print(y_batch.shape)
    p = np.argmax(y_batch, axis=1) # predict를 통해 나온 100 * 10 에서 행에서 가장 값이 큰 인덱스 반환
    print(f'y_batch: {y_batch.shape}')
    print(p)
    print(f't: {t[i:i+batch_size]}')
    accuracy_cnt += np.sum(p == t[i:i+batch_size]) # t에서 100개씩 짤라 p와 비교하여 맞은 개수를 갱신

print("Accuracy:" + str(float(accuracy_cnt) / len(x)))
```
